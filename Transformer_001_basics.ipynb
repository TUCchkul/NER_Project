```python
#Text Classification
classifier=pipeline("sentiment-analysis")
```

```python
classifier("I love my country")
```

```python
classifier(["]I do not love my country", "I did not like the movie that I watched last night"])
```

```python
review="""Rocky is an ambitious boy, who wants to die as the richest and most powerful man. 
From the streets of Bombay to the fields of KGF, he embarks on the mission.
 And his tale is being retold by a journalist who chronicled his life in the 1980s."""
```

```python
classifier(review)
```

```python
# Zero shot classification
```

```python
classifier_zero_shot=pipeline("zero-shot-classification")
```

```python
classifier_zero_shot(
    "This course is about Artificial Intelligence, here we will learn theory and practical both",
    candidate_labels=["politics","business","education"],
)
```

```python
# Text Generator
generator=pipeline("text-generation")
```

```python
generator("This course is about AI")
```

```python
generator("This course is about AI")[0]["generated_text"]
```

```python
generator=pipeline("text-generation", model="distilgpt2")
```

```python
generator("This course is about AI",
          max_length=40,
          num_return_sentences=3)
```

```python
generator("The movie was horrible",
          max_length=40,
          num_return_sentences=3)
```

```python
generator("<p>The movie was horrible <p>",
          max_length=40,
          num_return_sentences=3)
```

```python
generator("SELECT * FROM TABLE",
          max_length=40,
          num_return_sentences=3)
```

```python
unmasking=pipeline("fill-mask")
```

```python
unmasking("This course is about <mask> intelligence", top_k=3)
```

```python
NER=pipeline("ner", grouped_entities=True)
```

```python
NER("I am kulkirti and I work in Tu Chemnitz. I do live in Chemnitz")
```

```python
QA=pipeline("question-answering")
```

```python
QA(
    question="where do I work",
   context="I am sunny and I work in iNeuron Intelligence. I live in Bangalore."
)
```
